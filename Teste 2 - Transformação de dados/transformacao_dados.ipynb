```python
import pandas as pd
import tabula
import os
```

Escolhi trabalho com jupyter notebook esse teste, pois facilita a visualização das tabelas

```python
#pdf utilizado
file = "componente_organizacional_202108.pdf"
#cria pasta para salvar csvs
try:
    os.makedirs("csvs_salvos")
except:
    print("O diretorio ja existe")
    #faz o que acha que deve se não for possível criar
```

```python
#Leitura da tabela 30
tabela30 = tabula.read_pdf(file, pages = ('108'), multiple_tables = True)
print(len(tabela30))

display(tabela30[0])
```

```python
#formatação e visualização da tabela 30
tabela = tabela30[0]
nome_da_tabela = tabela.columns
tabela[[0,1]] = tabela[nome_da_tabela[0]].str.split(" ", expand = True,n=1)
tabela.columns = tabela.iloc[0]
tabela = tabela.drop("Código Descrição da categoria",axis=1)
tabela = tabela[1:]
tabela = tabela.set_index("Código")
display(tabela)
#Salva tabela no CSV
tabela.to_csv('csvs_salvos/'+nome_da_tabela[0]+".csv")
```

```python
#Leitura da tabela 31 e 32
tabela31_32 = tabula.read_pdf(file, pages = ('109','110','111','112','113','114'), multiple_tables = True)
#retira o numero de tabelas
print(len(tabela31_32))
```

```python
#tratamento para o começo da tabela 31
tabela_begin = tabela31_32[0]
nome_da_tabela = tabela_begin.columns
tabela_begin.columns = tabela_begin.iloc[0]
colunas = tabela_begin.columns
tabela_begin = tabela_begin[1:]
tabela_complete = tabela_begin
display(tabela_begin)
nome_da_tabela = nome_da_tabela[1]
print(nome_da_tabela)

```

```python
#tratamento para o resto da tabela 31 e a tabela 32, o final da tabela 31 e a tabela 32 ento na mesma pagina
for index,tabela in enumerate(tabela31_32):
    if index>0 and index<6:
        #tira primeiro elemento do titulo da tabela e organiza
        tabela.loc[-1] = tabela.columns
        tabela.index = tabela.index + 1
        tabela = tabela.sort_index()
        #define novo titulo
        tabela.columns = colunas
        #display(tabela)
        tabela_complete = pd.concat([tabela_complete, tabela])
    if index == 6:
        nome_da_tabela32= tabela.columns
        tabela[[0,1]] = tabela[nome_da_tabela32[0]].str.split(" ", expand = True,n=1)
        tabela = tabela.drop("Tabela de Tipo de Solicitação",axis=1)
        tabela.columns = colunas
        tabela = tabela[2:]
        tabela = tabela.set_index("Código")
        tabela = tabela.sort_index()
        tabela = tabela[:-1]
        tabela32 = tabela.to_csv('csvs_salvos/'+nome_da_tabela32[0]+".csv")
        display(tabela)

#Salva tabela 31 no CSV

tabela_complete = tabela_complete.set_index("Código")
display(tabela_complete)
tabela_complete.to_csv('csvs_salvos/'+nome_da_tabela+".csv")
```

Zipando os arquivos salvos

```python
import zipfile
```

```python
def zipdir(path, ziph):
    # ziph is zipfile handle
    for root, dirs, files in os.walk(path):
        for file in files:
            ziph.write(os.path.join(root, file))
```

```python
meu_nome = "Ricardo_Anchieta"
zipf = zipfile.ZipFile( 'Teste_Intuitive_Care_'+meu_nome+'.zip', 'w', zipfile.ZIP_DEFLATED)
zipdir('csvs_salvos/', zipf)
zipf.close()
```
